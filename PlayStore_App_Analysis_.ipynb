```python
#stripping size column and converting all to Megabytes
google["Size"]=google["Size"].map(lambda x: x.rstrip("M"))
google["Size"]=google["Size"].map(lambda x: str(round((float(x.strip("k"))/1024),1)) if x[-1]=="k" else x)
google["Size"]=google["Size"].map(lambda x: np.nan if x.startswith("Varies") else x)
google["Size"]
```

```python
#stripping Installs column of special characters
google["Installs"]=google["Installs"].map(lambda x: x.rstrip("+"))
google["Installs"]=google["Installs"].map(lambda x: "".join(x.split(",")))
google["Installs"]
```

```python
#stripping Price column of special characters
google["Price"]=google["Price"].map(lambda x: x.lstrip("$").rstrip())  
google["Price"]
```

```python
#converting data types to numeric
google["Size"]=pd.to_numeric(google["Size"])
google["Installs"]=pd.to_numeric(google["Installs"])
google["Price"]=pd.to_numeric(google["Price"])
google["Reviews"]=pd.to_numeric(google["Reviews"])
google["Last Updated"]=pd.to_datetime(google["Last Updated"])

```

```python
#checking data types
google.dtypes
```

```python
#dropping duplicates by duplicate app name
google = google.drop_duplicates(subset = "App", keep="first").reset_index()
google = google.drop(labels="index", axis="columns")
google
```

```python
#checking for na values
google.isna().sum()
```

```python
#dropping records with na values
google = google.dropna()
```

```python
#checking size of dataframe, na values, and datatypes
google.info()
```

```python
#resetting index
google = google.reset_index()
```

```python
#dropping extra index column
google = google.drop("index", axis="columns")
```

```python
#splitting genres into two seperate columns and assigning to dataframe
genres = google["Genres"].str.split(";", expand=True)
genres["Genre 1"] = genres[0]
genres["Genre 2"] = genres[1]
genres = genres.drop(labels=[0, 1], axis="columns")
```

```python
#stripping genre columns of special characters
genres["Genre 1"] = genres["Genre 1"].str.replace(" & ", "_")
genres["Genre 2"] = genres["Genre 2"].str.replace(" & ", "_")    
genres["Genre 2"] = genres["Genre 2"].str.replace(" ", "_")
```

```python
#joining counts of each genre from both columns
genres_count = pd.concat([genres["Genre 1"].value_counts(), genres["Genre 2"].value_counts()], axis=1)
genres_count = genres_count.fillna(0)
genres_count["Count"] = genres_count["Genre 1"] + genres_count["Genre 2"]
genres_count = genres_count.drop(labels=["Genre 1", "Genre 2"], axis=1).sort_values(ascending=False, by="Count")
genres_count
```

```python
#created new dataframe to create bins for price pie chart
goog_price = pd.DataFrame(google["Price"].astype('str'))
```

```python
#created % of app by price pie chart
goog_price["Price"]=goog_price["Price"].str.replace("0.0","0.99")
goog_price["Price"] = goog_price["Price"].astype('float64') 

price_apps = pd.cut(goog_price["Price"], bins=[0, 0.99, 1.99, 2.99, 3.99, goog_price["Price"].max()],
                   labels=['< $1.00   93.9%', '$1.00-1.99   1.4%', '$2.00-2.99   1.6%', '$3.00-3.99   0.8% ','> $4   2.3%'])
price_bin = price_apps.value_counts(normalize=True, sort=False)

fig_price_pie, ax_price_pie = plt.subplots(figsize = (9,4.5))
price_bin.plot(kind='pie', ax=ax_price_pie, radius=1.2)
ax_price_pie.set_ylabel('')
fig_price_pie.suptitle('% Apps by Price')
plt.savefig("price_percent.png", bbox_inches="tight")
```
